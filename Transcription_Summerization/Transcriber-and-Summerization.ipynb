# **Using Bart Model and Tokenizerfor generating Summary ,Extracting Key points and generating Title**

```python
from transformers import BartForConditionalGeneration, BartTokenizer

# Load the BART model and tokenizer
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')

def generate_title(text):
    # Encode the input text
    inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)

    # Generate the title
    title_ids = model.generate(inputs, max_length=15, min_length=5, length_penalty=1.0, num_beams=4, early_stopping=True)
    title = tokenizer.decode(title_ids[0], skip_special_tokens=True)
    return title

def extract_key_points(text, num_points=5):
    # Encode the input text
    inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)

    # Generate the summary
    summary_ids = model.generate(inputs, max_length=1024, min_length=100, length_penalty=1.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # Split summary into sentences and extract key points
    sentences = summary.split('. ')
    key_points = sentences[:num_points]
    return key_points

# Example usage
long_text = """

Welcome to my world. Welcome to my world. Want to go in. Miracles I guess. Still happen now and then. Extending to my heart. Leave who cares in my head. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. I'm not going to die. For most of human civilization, we believed that we were Earth's most dominant species. We believed that life could only exist on the surface of the planet. Well, after a certain point you have to wonder, what else were we wrong about? We have maps less than 5% of hollow Earth. But our ecosystems are linked in ways we never could have imagined. We are not two separate worlds. We're one. Kong is settling in to her worth. So the big guy likes his new home. We think so. His species was a very social one. It's not natural for Kong to be so alone. He's the last of his kind. Every day he searches for a family that he'll never find. What about Godzilla? Godzilla's on the surface. Kong is below. As long as they don't venture into each other's territories, we've got nothing to worry about. Humanity is lucky. Godzilla's fighting the battles we can't. The battle is over. The battle is over. The battle is over. The battle is over. The battle is over. The battle is over. The battle is over. The battle is over. The battle is over. The battle is over. Kong is setting in for the night. Looks like it was another rough one. That's the part that's gone yet. Maybe he was looking forward to date. Good luck with that. I hope he's up to some meter species romance. Well. Simple detected in sector five. That's contacting ground. That's contacting ground. Oh. Keeps doing that. More electrical anomalies? I don't think so, man. I haven't seen anything like this before. I don't know if you'd tell the director Hampton. I'll type it up. I'm so claseless in an anomaly. So mean. Tell the Italian military where one would be flattened if it worked for Godzilla. They can threaten him all they want. Just tell them to find another scapegoat. Director Hampton, what's up? Well, Outpost 1 is still a night here. And they seem to miss over. I want to take a look. Because it's from the sensor array? Yeah, we've been getting some pretty weird spikes. I mean, it could just be regular and deferring. Yeah, but it keeps happening. We should prep a server to him just in case. Okay. And what about Godzilla? Oh. He's still sleeping. Like a big angry baby. Let's hope it stays that way. I just need to make a soul right now. Yeah. It's a real thing. I need to make sure we can get the camera. Yeah. Everyone, eyes, you know, paper, all right? Let's make sure we circle the answer. Really? Yeah. Your daughter is intelligent, but she's not engaged. Her grades are slipping. She's not interacting with any of her friends. And these problems aren't going away, Dr. Andrews. They've been getting worse. Okay, I'm sorry. She's just still adjusting to our world. I understand that, but we have a number of students who are culturally displaced. Sure, but her culture is not displaced. She is her culture. She's not engaged. She's not engaged. Sure, but her culture is not displaced. She is her culture. She's the last living member of the EU we tried. We're just trying to reach her. I know. Trying the club. Why do you like camera? I'm not a<|sv|> I don't know. I'm going to hide behind you. Come on. For centuries, the ewe tribe were the guardians of Stol Island, the ancestral home of the Great Colm. But when shifting storm patterns destroyed the island, the ewe were annihilated with only a single member of the tribe having survived. A young girl found and rescued amid the punishing storm. Her fellow villagers were lost for the tuts, gone to history. Just like the Great Colms in Tess. Yes. Nothing on the ship. Perfect. I'm going to figure this out. Where will you be when the next tightening attacks? How will you protect the things and people that matter most to you? Today's podcast is brought to you by Gargantuan Insurance, the only insurance company that provides tight protection for your home and personal belongings. Because Godzilla may not care about your hopes and dreams. But Gargantuan does. Okay, tighten. We have got a button to attack on today's show. Recording in progress, go away. Yes, I just want to let everyone know that there is no way for us. Let's... I'm going to kill. Unbelievable. What? Hi, Bernie. I think you're going to be able to get the right answer. I'm going to kill. I'm going to kill. Hi, Bernie. I mean, Andrews, we met in Hong Kong. Oh, oh, yeah. Yeah, yeah, yeah, yeah. Hong Kong. Yeah, that was so long ago. Oh, no. God. Yeah. My hair was long. Cut your hair. Highlights, too. It's nice. I was wondering if I could get your help with something. Me. Uh-huh. Sure. Of course, yeah, come on in. Dr. Andrews. Okay. Yeah. Straight down that hall there to the left. Have a seat. What can I help you with? You know, me, Kasa, Azuoh. Okay, so the sequence on the right, this is a signal that's been interfering with our equipment and holo-worth. Okay. And these drawings by my daughter, Gia, and you're going to think that I'm completely insane. You know, the patterns are the same in the middle section. Yes. Exactly. Thank you. Look, our analysts, they don't know what to make of this. They're talking in circles. They're saying radio interference, coincidence, but I- I know that you think outside of the box. I didn't know who else to turn to. And they sent you here because- No, no one sent me. I can't hear you. You know, that's what's ironic. You know, monarch wouldn't confirm my crucial part in taking down mecha-Godzilla which cost me thousands of- hundreds of subscribers on my blog. Oh, and the trolls, the trolls online. They question every single thing that I put up, especially this one troll. This one troll named Gadorasthan64, you ever heard of them? No, I don't think so. Great because he's a trash bag, okay? I'm more qualified than half of those people that you have working over there. I knew what Gazilla was going to do before anybody would monitor. But this is why I am here. Bernie, I am here. Okay. Can you help me or not? That depends. Are you the new boss down at the mess? At what? The monarch, Hollow Earth Station, the M.H.S., the mess. Yeah, nobody calls it that. I run the Kong Research Division, but technically director Hampton is in charge of general operations. Well, director Hampton doesn't have a kid who can talk to Kong. You want to go to Hollow Earth? I'm so glad I didn't have to say it. I thought you were never going to ask, but I really do need proof. That would help me with my bonafides for my blog. And I knew that you would be the person that helped me do that. Okay, well if this happens and that is a big ifs, then monarch would need, you know, whatever you guys call it. Final cut on it. Oh, well, I'm sorry, final cut. No, no, no, no, no, no, that will destroy my journalistic integrity. No really true, good documentarian, give final cut. I could make you a producer. Okay, okay, final cut, final cut. Something is happening to my daughter and I have no idea how to fix it. You work out what's going on and I'll get you anything that you want, okay? That is a promise. You're going anything is a lie, but that's very that's very generous. And we have to go right now. What? Godzilla is on the move again. Okay, listen to me. If Godzilla is on the move, he senses a threat coming. We don't know what it is. Of course we're concerned, Senator. That's why we're monitoring the situation. What is it? And I am telling you, Senator, that Kong doesn't leave Halloward, which is why Godzilla has no reason to do that. What is it? And I am telling you, Senator, that Kong doesn't leave Halloward, which is why Godzilla has no reason to tell you. What's he doing up here? No idea. Just Godzilla, no. No, that we can tell. It was only a matter of time. And what happens if Godzilla starts hitting this way? And we all run like hell. So that's in place. Right, the good news is, it's just an infracted tooth. But luckily you've come to the best doctor in town. I got you. All right, concentration at 100% vital stable. Trapper, you are clear to swarm the damage tooth with replacement. Hey! Hey! Touchdown Trapper. Hey! Hey! Hey! Hey! Touchdown Trapper. Oh, fella. Do you have the breakfast? All right. Yeah, this sucker's ready to come out. All right, let's go. There she is. Trapper, thanks for getting down here so fast. Nice work. You are most welcome. So every day you get to climb inside a 300 foot 8 mouth. Is it? Yeah. Are you sure it's going to be strong enough? The tooth. Yeah. That's the same polymer composite that you use on the vehicle heat shields. You can jump through the rifle tower, I think. Well, good teeth aren't going to help if Godzilla senses the heat's up here. Well, you can't have a Titan with a tooth, hey, Kenyut. No. Don't worry. A couple of hours, you might be a bit groggy, but it'll be good to go back down. You know, when I first heard that you were going to be a vet, I thought, no, that is way too boring for Trapper. Then you go and become the weirdest vet in the world. Yeah, it's basically me and Dr. Doe Little. I set my animals a bigger. I thought finally someone has made him grow up. Nope. Someone tried that back in college, but didn't really work. Oh, yeah? Yeah, what was she said to me? Trapper, you are not a serious person. Oh, she sounds very smart. Yeah. Yes, she's got a PhD. Too, actually. How's the kid, Dei? He's up, struggling, trying to find a place in the world. Yeah, I know where that goes. Don't worry, she'll be all right. She's got a good mum. Oh, yeah. Yeah. Yeah. Yeah. Yeah. Yeah. Yeah. Yeah. Dr. Andrews, your guest has been asking for you. Please don't take this the wrong way, but can I just say that your research library has many glaring omissions. I had to call in some serious favors from my discord chat. Bernie, do you have something? Yes, I've got something. I do have something. Whatever signal you guys are picking up. Okay, we have seen similar patterns repeating overtime in each spike corresponding to the specific Titan event. Thank you, hold this, please. Thank you so much. Okay, check this out. See this? King D'Dora. First encounter with Godzilla. Here, discovery of Skull Island. All right, so what you're saying is that these signals are just random patterns. Yeah, it seems that way, yeah. And Godzilla's picking up on it. Oh, I can do you one better. Now, look, I've looked at that point of similarities in each waveform and I compared them to that pattern. Bernie, what are you saying? It's a distress call. It's a distress call. A psychic energy SOS. If your daughter is seeing these symbols, then something down there is calling for help. Regular sound waves, they can't propagate this far without decay. This is something that we have never encountered. And it's getting stronger. You got a waking earring? Here, it's telepathic alien language. Some sort of distress signal. Signal? From what? Or from who? Well, whatever it is. This is why Godzilla's on the move. We can't say for sure, but it seems that way. Well, we have to figure this out fast. The walls are already at our door. The government has been looking for an excuse to take over hall of worth operations. And the corporations are looking at chops. Yeah, because who wants to be rich and get rid of all the monsters? It may not be terrible, right? We better find the source before DC gets wind of this. Okay. I'm giving you full operational oversight. Take a small team, we kale, make it fast, and don't let anyone die. We'll be in the night out there before nightfall. You, uh, you feel like going for a ride? Do you never ask? Attention all ahead. Prepare heat three for hollow earth launch in T-minus five minutes. There's no response from the hollow earth thyprost. The radios must be done. Well, then that's our first stop. Copy that. Copy that. Waky, waky, everyone. Congs up. Okay. Hey, hey, the angel. Hey, come on, we had a video. Bernie, this is Insighting. It's a reconnaissance mission. Where? So I see how it is. When you need something, it's all. You're so smart, Bernie. You're so capable, Bernie. I'll subscribe to your Titan blog, Bernie. And then when I do it, I'll say, I'll just say, I'm going to do it. And then I'll just say, I'm going to do it. I'll just say, I'm going to do it. And then I'll do it. And then I'll just say, I'm going to do it. You're so capable, Bernie. I'll subscribe to your Titan blog, Bernie. And then when I deliver it, you toss me inside. Okay, look. This could be dangerous. Come on. I have no stranger to danger. I didn't mean to rhyme that time. Look, you owe me. Okay? I hope crack this thing. This could be like last time when we save the world. Remember? Hmm? Mikael? Yeah. Propose for one more. Thank you, Dr. Andrew. Dr. Andrews, you will not regret it. You're kidding, right? I'm the plus one. Hello. Oh! Oh, we're online. We are going to Hollywood. I am going to Hollywood. Oh, my God. What am I doing? So is there like an inflight briefing that I should know about? You know, there is. There's no parachutes, no flirtation devices. You touch under this stuff, you lose your hand, and if you die, make sure it's outside my vehicle. Good briefing. Great briefing. Don't take me keel too personally. Aggression is his love language. All right, the keel. Let's punch it. All right, hatheEO, let's punch it. Did you see when I fell in a knee injustice Forgive me. We just took nothing from him Let's finish. Oh Just try not to swallow your tongue Try not to swallow my tongue at all. I don't exactly know what that means On your count Dr. Andrews Oh I'm turning into Oh So systems active systems stabilized all systems good doc as ever with it. You're right never better. Yeah Bernie you good Nice one oh Looks like Kong is headed home. How far are the apples? 100 clicks still not answering comms Possibility alert Predators 9 o'clock where what that was a verta scenes Mickey. Oh, they're called verta scenes from a their problem They're totally harmless unless you get rolled up and then they pack about a lightning bolt worth a bio electricity Oh Predators know to stay well away which makes them the perfect chef-owned if you happen to have I call by a mimicry camera flash which we do Oh Talk to me. It's not good. Godzilla just made lamb fallen France ground team is five clicks out Oh We got picture oh my god. It's a nuclear facility Oh That atmospheric radiation levels are dropping he's absorbing it. Okay. He's charging for something Oh A ends We're coming up on outpost one. Oh god. Oh no, that doesn't look good. No, does it? You've seen the stock? Yeah, yeah, I'm seeing it. Oh. Well, there's no heat signatures. There's so much for survivors. God, I have some respect. These people had families. Who could have done this? I don't understand. We've had this outpost for years without incident. So why attack now? What changed? Here, Doc. There's a camera here. You reckon you can pull any footage off of that? Oh, shit. You think home to that? Oh, that's our con. That's something else. Yeah. Loose. Get off the ground before that something else comes back. Saddle up. Loose tossing. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. You getting anything off earth, eh? Yeah, I'm just... Oh, my God. What? Here. Take a look. It looks like an ape, but that's impossible. What the bloody hell is that? I have no idea. Mikael! Mikael! Stop working on that! Don't look! Don't look! Don't look! The hell was that, Mikael? Calm down. Calm down. I'm almost at the signal. The same electrical disturbances were happening at the outpost. The system's stabilized. Oh, come on. Come on. Oh. She says it's coming from that rich point. Well, let's take a look. Okay, it's close to the camera. The Hollow Earth. A world untouched by mankind. Is this how Neil Armstrong felt when he stepped on alien soil? Is this one small step for me and one giant leap? Please. Are you kidding me? Write down the lens. It's a documentary. You want to be natural. Just walk around. All right. Yes. The Hollow Earth. A world. All right, guys. That rich, you pointed out is about two miles straight ahead. I'm getting a vibe out here. Oh, look at that. That must be tasty too. Keller. Only female wasky toes drink, lad. So they can grow their eggs. She needs it all in me. Wait. What? Let's all hold up. Not this way. What? Why? Now we got to take a detour to something here. Where here? Like here, here? No, I'm not picking anything up. No, something doesn't smell right. It's not just you, Mekil. It's my life. It's like that rotting flesh. It's probably just the carcass I've went to. We keep moving. Mekil, hold up. Listen, pal. I'm gonna trust my state, the art thermal. Over your ace ventura, hippie, deppy, six cents, all day long. You got me? There's no need to call people names. That's for you, freak show. If you don't shut it, I'm gonna take that camera. I'm gonna shove it right up your ass. I'm a charge. We move. Why don't we just listen to somebody? I said I'm a charge. I'm a charge! God! All right, we should be all right. Those things don't reach that far. He was just eating by a tree. What's he doing, Doc? I'm trying to call the base for excursion. One minute he was there, ready, he was right in my face, and then a tree out of nowhere. No, man, no, that's heavy. That's the natural world, man. Red in tooth and cool. Tennyson, you just watched a man get devoured by a topiary nightmare, and you want to quote Tennyson down here. Yeah, man. In your all about it. Let love class brief, man. Come here. You know, humans, we think we took ourselves off the food chain, but maybe. Maybe the Titans are here to remind us. I think that there is something seriously wrong with you. Okay, okay, okay. It's not working. We better move on. We'll move on later. See ya, come on. Come on, back. We do. He was eaten by a tree. I don't want to die like that. Not by a tree. No, we can't get any closer than this. God's illness burning whole radiation. Whatever he's headed for has a really pissed off. This is base calling for being alpha over. Fifteen alpha responding. God's illness current trajectory is showing him heading towards the activation. What Titans are in the area? Looks like he's hunting Teamot. Classified file 324. Titanous Teamot. Titan 19. Extremely aggressive. Territorial. Weaponizing her body of razor sharp scales. The monster's arctic layer sits in the direct path of electromagnetically charged solar winds, making it the largest stockpile of energy on earth. If Godzilla takes down Teamot, he'll supercharge. This ruins human civilization. Down here. Hey. Oh. Yeah. The markings are older than the ones on skull and centuries older. But the architecture is consistent with the e-weep. This is remarkable. Wow. This is remarkable. No, that's who he is. There's no moss on these ones. No. A mullsora. Coin of the monsters in a hollow earth. Our lineage must go back further than we thought. Yeah. Yeah. No, no, no. No, no, no. No, no, no. What is that? Oh. Hey. Oh, one, one, one, one, one. This looks like some kind of an irrigation system. I mean, this predates everything. This botany is the lamirians. These guys were the first, man. And certainly he's going. Sorry. Yeah. Great modernist, Golly. Look at her. Agent Haley and astronauts. Yeah. Look at this here. The water's running underneath. And this is so canny. Yeah. I'm kind of bioelectric, hum. Oh. It's like camouflage. So, see me doing that? She said we're doing that, right? She said we're doing that, right? She said she. No, no, no. No, no, no. This must be the source of the signal. Oh, he's going to check this out. Okay. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Fifteen hours of reporting in place. We're entering TMI's domain. Something is off. TMI's energy is distorting our radar. We don't have a visual gun. Take us forward. Sallet running. Five knots. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. He's gone to TMI's layer. Wait. Wait. Wait. Wait. Wait. What was she saying? She says we're not alone. I knew. I knew. I felt something. Come on, man. Oh. Oh my god. Oh. Oh. Oh. Oh my god. What? It's an entire e-wiscivilization protected within all of our... These structures, they look like they've been carved from some enormous quartz crystal that must be their energy source. I'm feeling the energy but let's just keep it together. Oh my god. It's an undocumented vortex. It's like several of them actually. Ruts to the surface. These could lead all over the earth. Oh, wow. You're so good. You should make eye contact with them. It's considered polite. Yeah, this is not working for me? Oh. How is that communicated? Telepathically. No, no, no, no, no, no. You can't take her off. Hey, hey, don't put it up. Wait, no, no, no, no, no. What's she taking her? Hey, hey. Okay. Oh my god. Oh my god. Oh my god. Oh my god. Oh my god. Let's go. Let's go. Whoa. Oh my god. Oh my god. Oh my god. Oh my god. Oh my god. So what they said now is SOS. And then what? Godzilla hears it? They're walking up. And that's always for what? They already live in a nightmare monster hellskid. What could possibly scare them? Hey, she can understand everything you say. She understands me. And a nightmare monster hellskid is a wonderful place to raise a family. Whoa. Oh my god. Oh my god. Oh my god. Oh my god. Oh wow. This whole room is covered in a waste of time. It says that in the beginning, hollow earth lived in harmony with the surface world. The Titans were the guardians of nature. And the great apes became the protectors of humanity. But a great evil threatened the peace. A powerful and ruthless ape desperate to conquer the surface world led his tribe into war against the one they call the monster who led his star. Godzilla. There's more. The war with the apes nearly destroyed Godzilla. But after a great battle, he imprisoned the apes in a fiery realm of hollow earth. Their false king remains obsessed with reaching the surface. The ewe call him the scar king. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Oh. Trapped within their subterranean realm, the scar king harness a terrible power, the ancient type, Shima. He controls her with pain, her power covered the earth in the last ice age. We brought Kong down here. He's been looking for others like himself going deeper into hollow worth and getting closer to their prison. The ewe must have known it was just a matter of time and that's why they've been calling for help and that's why God's illness changing. He's carrying up a wall, sir. He barely survived last time. What's Godzilla going to do on his own? He won't be on his own. The ewe believe that at the end of the world, one of their own will return. An awakened Mothra, defender of the ewe and ancient ally of Godzilla. They believe that their savior will be an ewe from Skull Island. That seems like a lot to pull on one kid. This is on our air team to base tracking of our Godzilla's location. We've closed the shipping lanes within 300 miles. Classmarading syndicate, he's absorbing everything in there. You know, I made a deal with myself when I became a mother. I said to myself, I don't care what you have to sacrifice, what you have to give up. You will do right by that little girl. I just never thought that the thing I'd be giving up would be her. You don't know that. Really? Trapper before we came here she said that she didn't belong anywhere. This is the life that she wants then I have to give her up. What a funny turn does everything feel lighter down this? Yeah, do it, sir. Oh, they too. That must weigh a ton. Yeah, see, I don't know how they can do it, but somehow they're able to move the pair of mispositions, which disrupts the gravitational pull down here and that's how they run their city. An entire civilization built up the manipulation of gravity. God, I wish I was streaming this log. What are you going to do with that, pun? Look, I know what you're thinking, but you don't understand. Okay, through what I've been through seeing what I've seen and have absolutely nobody believe you. This place, this place is special, man. This magic. How long do you think it'll stay like that? It's not posting that about, hmm? Like you look at any isolated tribe or community, how many of them survive contact with the outside world? You know, I think it's brilliant, eh? You already saved the world once. I can't say that away from you. Could you say that all again? The last part, so I have it on footage? What? Oh, well, we have to find out what that was, right? Oh, my God. Is that a mini-cong? How did they find this place? It must have sent you out. That arm looks bad. That's worse than bad. That's rough fun. I better go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have to go take a little bit. I'm going to have our time really fast. I'm are going to have our time really fast. around. Roda, detecting all truth energy signatures. Oh my god. He's changed. I know, I know. Nobody likes a dentist. Just get a look at that armpit guy. Trapper, he looks really hurt. Yeah, it's not great. I'm seeing nerve damage. A significant tissue loss. So what are we saying? He's not gonna last a day down here with only one arm. Yeah, communication to the surface is still out. If I can get back to the outpost, I think there might be some supplies that we can use. I don't think a massive cast is gonna solve the problem. I wasn't talking plaster of Paris. I'm thinking more along the line to... Project Powerhouse. You can't be serious. Oh, I'm deadly serious. That prototype was almost finished when they pulled the funding. All right, well, what are you still doing here? I love you, too, Doc. Project Powerhouse, what is that exactly? After mechagotzilla, we realized that there were some threats that even calm couldn't face. So we started working on some minor augmentations. Oh. We were kind of augmentations. Making one of the strongest titans even stronger didn't go over so well, so we got shut down. Lucky for us, the prototype had already been transported to Hollywood for testing. It stored in the armory at outpost one. The other column would be damaged. What do you think of that? That's one hell of an augmentation there, Trapper. Only the best for Mr. Kong. I'm not sure if I can do this. All right, ready to go. Those injections should heal the frost by up nicely. All right, there he goes. He's moving. We got this. It's a bit rough from ready, but it should hold. It looks good, Trapper. Damn good. There we go. See if he can love it, or he's going to rip it off or just take it. Ah! I think he loves it! Hey! This is scarking. Hey, hey! Why are you telling me? Where's he going? Godzilla won't come down here unless Kong brings him. Okay, the last time those two met up, it was almost the end of Kong. He's taking a hand of a wrist. Kong invades his turf and start a war. Yeah, that's a suicide mission. But if Kong draws Godzilla down here, they can make their stand in Hollywood. And there's a chance that Godzilla can stop scarking and Shimo from reaching the surface. The first time he's been here, he's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. He's been here. We are absolutely certain that she can pull this off, right? It's why Jaya was called here. Only an Ewe from Skull Island can awaken Mothra. And Mothra is the key to Godzilla protecting this place, and its pathways to the surface. She's going to be here. Uh-huh. She's going to be here. Uh-huh. She's going to be here. Uh-huh. She's going to be here. Uh-huh. Uh-huh. Uh-huh. Uh-huh. Uh-huh. Uh-huh. Uh-huh. Uh-huh. Uh-huh. Arhhhh... This distance is from the ship that individual pose for this game. Does that make sense? Whoa. Why, wait a minute. Drinks, but how does that make sense? There's a store on to the left. A moment. Hey, she's evacuating the villagers. Gravity, gravity, the pyramids and gravity, yeah, I got it, I got it now. Tell me on the way. Come on, come on. While we are throwing random shit at the wall, permission to take the heat. Rally some reinforcements. To the surface? No, there's no time. No, this idea is way weird, isn't it? No, no, no. No, no. No, no, no. No, no. No, no, no. No, no, no. No, no. No, no. No. All right, we're underneath the pyramids, now one. Gravity, ewe technology. What the chemical reaction this liquid metal becomes in engineering mechanism, forcing the two electromagnetic pyramids together, causing an anti-gravity shockwave, but only for a few minutes or maybe less, which should keep those things from the portals. After that, everything that is up will come crashing. We're here. We're here. We're not going to make it this long enough time. We need something to slow them down. What else can we do? This bird has seen amazing season. No, no, no. No way. I've got to do it now. You're carried away. I'm going to dissuade them now. Back up is beautiful people. Open the way up. I'm going to die. I'm going to die. I'm going to die. No, the engines are out. The gravity surface has taken out the power battery thing. Yes, yes! She's protected the portals. I'm going to die. I'm going to die. Oh, God! Oh, God! The portals are far from here. It's here. I'm going to die. I'm going to die. I'm going to die. Oh, God! Oh, God! Oh, God! Oh, God! Oh, God! Oh, God! Oh, God! Oh, God! Oh, God! Oh, God! Oh, God! Oh, God! Oh, God! Oh, God! Oh, God! James, come. Don't wanna do this again, Mo. Help! I was somewhere out of here. Look, I... I know that this place... is like home. And I want you to know that whatever you're here decide I... What am I talking? I assume you want to stay. This place could be your home. I... Could you... You want it? Yeah. I'm sitting out about you. Every day my mind is all around you. Thank you. Thank you. You guys in the world. It's kind of hard to say. Looking out of my lonely room, day after day. I'm going to be home. Baby, we can't survive. I can't fight out to you. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home. I'm going to be home.
"""
title = generate_title(long_text)
key_points = extract_key_points(long_text)

print("Title:", title)
print("Key Points:")
for i, point in enumerate(key_points, 1):
    print(f"{i}. {point}")
```
